# Visualizations: Mapping Gap Events

### Import Packages and Data

```python
#Hack to fix missing PROJ4 env var
import os
import conda
conda_file_dir = conda.__file__
conda_dir = conda_file_dir.split('lib')[0]
proj_lib = os.path.join(os.path.join(conda_dir, 'share'), 'proj')
os.environ["PROJ_LIB"] = proj_lib
from mpl_toolkits.basemap import Basemap

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in data
gap_notreg = pd.read_csv('../data/gap_notreg_off_clusters.csv')
```

```python
# Break into individual dataframes
cluster0 = gap_notreg[gap_notreg['off_cluster'] == 0]
cluster1 = gap_notreg[gap_notreg['off_cluster'] == 1]
cluster3 = gap_notreg[gap_notreg['off_cluster'] == 3]
```

```python
# Create one dataframe with main clusters

top3 = gap_notreg.loc[(gap_notreg['off_cluster'] == 0) | (gap_notreg['off_cluster'] == 1)| (gap_notreg['off_cluster'] == 3)]
```

```python
# Transform data to numpy arrays for use in map
lat = top3['off_lat'].to_numpy()
lon = top3['off_lon'].to_numpy()
pos = top3['positions_per_day'].to_numpy()
hrs = top3['gap_hours'].to_numpy()
vessclass = top3['vessel_class'].to_numpy()
clusters = top3['off_cluster'].to_numpy()
```

## Global Map

```python
# Draw the map background
fig = plt.figure(figsize=(15, 15))
m = Basemap(projection='mill', resolution='h', 
            llcrnrlon=-100, llcrnrlat=-20, # lon, lat lower left
            urcrnrlon=200, urcrnrlat=80) # lon, lat upper right


m.shadedrelief()
m.drawcoastlines(color='gray')
m.drawcountries(color='gray')
m.drawstates(color='gray')

# scatter lat/lon data, with color reflecting cluster
m.scatter(lon, lat, 
          latlon=True,
          s=1,
          c=clusters,
          cmap='Reds',
          alpha=0.5)

plt.title('Largest 3 Vessel Clusters from DBSCAN Model');
```

## Cluster 0 Map

```python
# Create the map for Asia

fig = plt.figure(figsize=(15, 15))
m = Basemap(projection='lcc', resolution='h', 
            lat_0=30, lon_0=125, 
            width=6E6, height=6E6)
            

m.shadedrelief()
m.drawcoastlines(color='gray')
m.drawcountries(color='gray')
m.drawstates(color='gray')

# scatter lat/lon data, with color reflecting cluster
m.scatter(lon, lat, 
          latlon=True,
          s=pos,
          c=np.log10(hrs), 
          cmap='Reds',
          alpha=0.5);
```
